### Predictive Analytics

We would import LinearRegression Model and RandomForest Model
But our strategy is to take the correlated features (as obserevd from our exploratory analysis) and try and predict the total students' number 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    educ_data2['Num Tchr Full Time'],
    educ_data2['Num Students Total'],
    c='black'
)
plt.xlabel("Total Students")
plt.ylabel("Num of Full Time Teachers")
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    educ_data2['Num Students Female'],
    educ_data2['Num Students Total'],
    c='black'
)
plt.xlabel("Total Students")
plt.ylabel("Num Students Female")
plt.show()
```

What we have here is clearly a strong correlation. Let's see what we can make of it.

```python
X = educ_data2['Num Students Female'].values.reshape(-1,1)
y = educ_data2['Num Students Total'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    educ_data2['Num Students Female'],
    educ_data2['Num Students Total'],
    c='black'
)
plt.plot(
    educ_data2['Num Students Female'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Female Students")
plt.ylabel("Students Num")
plt.show()
```

```python
X = educ_data2['Num Students Female']
y = educ_data2['Num Students Total']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score



features = educ_data2['Num Students Female'].values.reshape(-1,1)
labels = educ_data2["Num Students Total"].values.reshape(-1,1)

# split the data into training and testing
train_feats, test_feats, train_labels, test_labels = tts(features, labels, test_size=0.2)


# Random Forest Classifier
clf = RandomForestClassifier()

# print the details of the Classifier used
print("Using", clf)

# training
clf.fit(train_feats, train_labels)

# predictions
predictions = clf.predict(test_feats)
print("\nPredictions:", predictions)

print (accuracy_score(test_labels, predictions))
```
